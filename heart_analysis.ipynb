# Chest Pain Analysis

```python
data.columns
```

```python
sns.countplot(x='cp',hue='target',data=data)
plt.legend(labels=['No-disease','disease'])
plt.show()
```

# Comparing Resting Blood Pressure as per Gender

```python
a = sns.FacetGrid(data,hue = 'sex',aspect = 4)
a.map(sns.kdeplot,'trestbps',shade = True)
plt.legend(labels = ['Male','Female'])
plt.show()
```
